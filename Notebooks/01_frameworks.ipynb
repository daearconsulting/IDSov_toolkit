## The Four-Framework Stack

This toolkit implements four complementary frameworks that together
provide comprehensive governance for Indigenous data:

| Framework | Governs | Origin |
|---|---|---|
| OCAP® | Who owns and controls the data | First Nations Information Governance Centre |
| CARE | Ethical obligations to communities | Global Indigenous Data Alliance |
| FAIR | Technical data management standards | GO FAIR Initiative |
| IEEE 2890-2025 | Provenance documentation | IEEE Standards Association |

These frameworks each addresses a different dimension
of data governance.

## OCAP®: Ownership, Control, Access, Possession

OCAP® was developed by the First Nations Information Governance Centre
to address a specific problem: researchers were collecting data about
Indigenous communities, publishing it, and the communities had no say
in how it was used or whether it was accurate.

The four principles address four distinct rights:

**Ownership**: A community or group owns information collectively in the
same way that an individual owns their personal information. Ownership means 
the community has the same rights over their collective data that an
individual has over their personal data.

**Control**: The rights of First Nations peoples and their representative
bodies to control data collection processes in their communities. Control
is the right to participate meaningfully in decisions that affect the community.

**Access**: First Nations peoples must have access to data about themselves
and their communities, regardless of where that data is held. If a
federal agency collected data about a community 30 years ago, the community
has the right to access and review that data.

**Possession**: The mechanism through which ownership can be asserted and
protected. A community cannot exercise ownership or control if they do not
physically hold a copy of the data. Possession without ownership is
insufficient, a database administrator holds data without owning it.

Reference: https://fnigc.ca/ocap-training/

```python
# OCAP® in practice, or what each principle means for our repos

ocap_examples = {
    "Ownership": [
        "Tribal-collected field data is owned by the Nation, not the researcher",
        "Federal data describing Tribal lands carries moral ownership even if legal ownership is ambiguous",
        "Publications about Tribal territory should credit the Nation, not just the research team",
    ],
    "Control": [
        "The Nation approves data collection protocols before fieldwork begins",
        "The Nation can halt data collection at any time",
        "The Nation has input on research questions and analytical methods",
    ],
    "Access": [
        "All data and findings are returned to the Nation in accessible formats",
        "Community members can request access to data about their territory",
        "Technical outputs include plain-language summaries",
    ],
    "Possession": [
        "The Nation holds a physical copy of all data; not dependent on researcher's server",
        "Tribal-collected data is gitignored and stored locally under Tribal control",
        "Data governance agreements survive the end of a research relationship",
    ],
}

for principle, examples in ocap_examples.items():
    print(f"\nOCAP® {principle}")
    for ex in examples:
        print(f"  • {ex}")
```

## CARE: Collective Benefit, Authority to Control, Responsibility, Ethics

CARE was developed specifically to address gaps in FAIR, which governs
technical data management but says nothing about ethical obligations to
the communities data describes.

**Collective Benefit**: Data ecosystems should be designed and function
in ways that enable Indigenous peoples to derive benefit from the data.
The data should create value that flows back to the community.

**Authority to Control**: Indigenous peoples' rights and interests in
Indigenous data must be recognized and their authority to control such
data be empowered. Authority is ongoing governance.

**Responsibility**: Those working with Indigenous data have a
responsibility to share how it is used and to support the capacity of
Indigenous peoples to govern their own data.

**Ethics**: Indigenous peoples' rights and wellbeing should be the
primary concern at all stages of the data lifecycle, including design,
generation, analysis, and dissemination.

Reference: https://www.gida-global.org/care

## FAIR: Findable, Accessible, Interoperable, Reusable

FAIR governs technical data management standards. It emerged from
scientific data infrastructure concerns about how you make data usable
across institutions and over time?

FAIR and CARE are often presented together precisely because FAIR is
necessary but not sufficient for Indigenous data governance. A dataset
can be fully FAIR (publicly findable, technically accessible,
interoperable, reusable) while completely violating CARE and OCAP®.

**Key tension:** FAIR's "Accessible" principle could be interpreted to
mean all data should be publicly accessible. CARE's "Authority to Control"
principle means some data should be restricted. The toolkit resolves this
by applying FAIR to the data that is meant to be shared, and OCAP® plus
CARE to determine what should be shared in the first place.

Reference: https://www.go-fair.org/fair-principles/

## IEEE 2890-2025: Provenance of Indigenous Peoples' Data

IEEE 2890-2025 is the first international technical standard specifically
for Indigenous data. It establishes that provenance records for Indigenous
data must document not only technical lineage (the standard W3C PROV
elements) but also governance lineage:

- **Whose land** does the data describe?
- **Under what treaty** authority was it collected?
- **Who has stewardship** obligations?
- **What governance frameworks** apply?

This standard operationalizes OCAP® and CARE into a machine-readable
format that can be attached to any dataset.

Reference: https://standards.ieee.org/ieee/2890/10318/
