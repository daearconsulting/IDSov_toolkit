## Sovereignty-first data collection

The most effective time to implement data sovereignty is at the beginning
of a data collection program before any data exists. Retrofitting
governance onto an existing dataset is possible but will require a lot more time and effort.

This notebook covers:
- Designing a data collection protocol with OCAP® built in
- Field template standards compatible with federal databases
- Classification schema for mixed-custody datasets
- The "data lifecycle" from collection to stewardship

## Data classification in practice

Every piece of data collected should be assigned a classification tier
at the moment of collection. The tier determines:
- Where the data is stored
- Who can access it
- Whether it can be committed to version control
- What sharing permissions apply

````python
# Show the classification tiers from the nation config
cfg = ctx.config
print(f"DATA CLASSIFICATION SCHEMA {cfg.name}")
for tier in cfg.classification_tiers:
    print(f"\n  [{tier['id']}] {tier['label']}")
    print(f"  {tier['description'].strip()}")
    if tier.get("examples"):
        print(f"  Examples: {', '.join(tier['examples'])}")
````

## The data lifecycle under OCAP®

```
DESIGN          Community questions with Nation approval
    
COLLECT         Field staff collect so Data stays local (gitignored)
    
PROCESS         Analyst processes - Provenance attached - Classification verified
    
REVIEW          Nation reviews findings and approves or modifies
    
SHARE           Nation authorizes sharing and Sovereignty statement attached
    
STEWARD         Nation holds copy and researcher access governed by MOU
    
ARCHIVE         Long-term stewardship plan in place
```

At every stage, the Nation retains control. The researcher is a
technical contributor, not the data owner.
