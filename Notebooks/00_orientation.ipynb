## What is this toolkit?

The Tribal Data Sovereignty Toolkit is a generalizable framework for
implementing Indigenous data sovereignty principles in Earth data science
workflows. It provides:

- **Python modules** that make sovereignty a built-in feature of data
  pipelines, not an afterthought
- **A notebook teaching series** that explains the governance frameworks
  and demonstrates their application
- **Document templates** for data governance agreements, provenance records,
  and compliance checklists

The toolkit was developed by Daear Consulting, LLC in the context of Earth
data science work with Oceti Sakowin Nations, and is designed to be
adapted for any Tribal Nation.

## Who is this for?

**Practitioners** (Tribal natural resource staff, data managers):
Start with `templates/` the document templates work without any coding.
Then read notebook 04 (Collection Design) for field data governance.

**Researchers** (scientists working with Tribal data):
Start here (Notebook 00), then work through notebooks 01–03 (Frameworks, Provenance,

Indigenous data.

**Students and educators** (courses, hackathons, workshops):
Work through the full series 00–06. Notebook 05 (REDS Workflow) is
designed for teaching responsible Earth data science.

## Quick Start

The fastest way to use the toolkit is three lines of code:

```
from toolkit.sovereignty import SovereigntyContext

ctx = SovereigntyContext.from_config("config/nation.yaml")
ctx.print_acknowledgment()
```

Everything else derives from the `SovereigntyContext` object.

## Adapting for a new Nation

1. Copy `config/nation_template.yaml` to `config/nation.yaml`
2. Edit every field (takes about 20 minutes)
3. Run: `toolkit/governance.py --config config/nation.yaml --validate`
4. All notebooks, templates, and modules auto-populate from your config

See notebook 06 (Adaptation Guide) for a complete walkthrough.

````python
# Load the default config (Oglala Lakota/Pine Ridge)
# Replace with your nation.yaml when adapting for another Nation

from pathlib import Path
config_path = REPO_ROOT/"config"/"nation_template.yaml"

if config_path.exists():
    from toolkit.sovereignty import SovereigntyContext
    ctx = SovereigntyContext.from_config(config_path)
    print(f"Loaded config for: {ctx.config.name}")
    print(f"Territory: {ctx.config.territory}")
    print(f"Active frameworks: {ctx.config.active_frameworks}")
else:
    print(f"Config not found at: {config_path}")
    print("Copy config/nation_template.yaml to config/nation.yaml and fill it in.")
````
