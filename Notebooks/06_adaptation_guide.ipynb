## Adapting the toolkit for a new Nation

This notebook walks through every step of adapting the toolkit for a
Tribal Nation other than the default (Oglala Lakota/Pine Ridge).
It takes approximately 30 minutes to complete the configuration.

**For Nation-specific consulting assistance**, contact Daear Consulting, LLC.
The core toolkit is generalizable; Nation-specific customization 
integrating existing governance codes, Local Contexts TK Labels.
Tribal data policies is a consulting engagement.

````python
import sys
from pathlib import Path

# Walk up to find repo root
REPO_ROOT = Path().resolve()
while not (REPO_ROOT/"config"/"nation_template.yaml").exists():
    if REPO_ROOT.parent == REPO_ROOT:
        REPO_ROOT = Path().resolve()
        break
    REPO_ROOT = REPO_ROOT.parent
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

print(f"Toolkit root: {REPO_ROOT}")
print("Starting adaptation guide.")
print("You will need:")
print("  1. Nation name and territory")
print("  2. Relevant treaty citation")
print("  3. Name of the Tribal department responsible for data governance")
print("  4. Bounding box for the territory (lat/lon)")
````

````python
# Step 1: Copy the template
import shutil
from pathlib import Path

template = REPO_ROOT/"config"/"nation_template.yaml"
target   = REPO_ROOT/"config"/"nation.yaml"

if target.exists():
    print(f"nation.yaml already exists at: {target}")
    print("Edit it directly or delete it to start fresh.")
else:
    shutil.copy(template, target)
    print(f"Copied template to: {target}")
    print("Now open this file and fill in every field.")
    print()
    print("Minimum required fields:")
    print("  nation.name")
    print("  nation.territory")
    print("  provenance.treaty_name")
    print("  provenance.treaty_status")
    print("  stewardship.department")
````

````python
# Step 2: Validate your config
from toolkit.sovereignty import SovereigntyContext
from pathlib import Path

nation_config = REPO_ROOT/"config"/"nation.yaml"

if not nation_config.exists():
    print("nation.yaml not found. Complete Step 1 first.")
else:
    ctx = SovereigntyContext.from_config(nation_config)
    errors = ctx.config.validate()
    if errors:
        print("Validation issues found:")
        for e in errors:
            print(f"  ! {e}")
        print("\nFix these before proceeding.")
    else:
        print(f"Config valid.")
        print(f"  Nation    : {ctx.config.name}")
        print(f"  Territory : {ctx.config.territory}")
        print(f"  Treaty    : {ctx.config.treaty_name}")
        print(f"  Frameworks: {ctx.config.active_frameworks}")
````

````python
# Step 3: Test the full acknowledgment
if nation_config.exists():
    ctx.print_acknowledgment()
````

````python
# Step 4: Register Nation-specific data sources
# The built-in source registry covers common federal datasets.
# Add Nation-specific sources here.

if nation_config.exists():
    ctx.register_source("tribal_water_quality", {
        "name":    "Tribal Water Quality Sampling Program",
        "steward": f"{ctx.config.name} Natural Resources Department",
        "license": "Tribal: governed by OCAP®",
        "url":     "data/raw/ (local only)",
        "note":    (
            f"Water quality samples collected by {ctx.config.name} staff. "
            "Under OCAP® principles, the Nation retains full ownership "
            "and control. Never publish without explicit Tribal authorization."
        ),
    })
    print("Custom source registered: tribal_water_quality")
    print()
    print("You can now use this source key in:")
    print("  ctx.attach_provenance(gdf, source_key='tribal_water_quality')")
    print("  ctx.generate_citations(['tribal_water_quality'])")
````

## Connecting to other repos

Once your `nation.yaml` is configured, connect the toolkit to your
analysis repos by adding this to their setup cells:

```python
import sys
sys.path.insert(0, "/path/to/tribal-data-sovereignty-toolkit")

from toolkit.sovereignty import SovereigntyContext
ctx = SovereigntyContext.from_config("/path/to/config/nation.yaml")
ctx.print_acknowledgment()
```

This replaces the individual `sovereignty.py` files in each repo
with a single canonical source.
