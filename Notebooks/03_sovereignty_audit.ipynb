## What is a sovereignty audit?

A sovereignty audit systematically evaluates a dataset or analysis product
against the four governance frameworks. It produces a scored report that:

- Identifies governance gaps before publication
- Guides data collection design from the start
- Demonstrates compliance to funders and reviewers
- Tracks improvement over time

The audit covers 27 questions across OCAP®, CARE, FAIR, and IEEE 2890-2025.
Each question is answered: **yes/no/partial/na**.

```python
import sys
from pathlib import Path

# Walk up to find repo root
REPO_ROOT = Path().resolve()
while not (REPO_ROOT/"config"/"nation_template.yaml").exists():
    if REPO_ROOT.parent == REPO_ROOT:
        REPO_ROOT = Path().resolve()
        break
    REPO_ROOT = REPO_ROOT.parent
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

print(f"Toolkit root: {REPO_ROOT}")
from toolkit.sovereignty import SovereigntyContext
from toolkit.audit       import SovereigntyAudit
ctx = SovereigntyContext.from_config(REPO_ROOT/"config"/"nation_template.yaml")
print("Ready.")
```

```python
# Create an audit for a specific dataset
audit = SovereigntyAudit("White River Streamflow Analysis for Pine Ridge 2024")

# Print all available questions for reference
print(audit.questions_list()[:1500])
print(f"\nTotal questions: {len(audit.questions)}")
```

```python
# Respond to questions
# In practice, respond to all questions that apply to your dataset

audit.respond("OCAP-O1", "yes",
    notes="Data governance MOU signed with Tribal Natural Resources Dept")
audit.respond("OCAP-O2", "partial",
    notes="MOU covers data sharing but publication rights not explicitly stated")
audit.respond("OCAP-C1", "yes",
    notes="Tribal Council approved watershed monitoring program")
audit.respond("OCAP-C2", "yes")
audit.respond("OCAP-A1", "partial",
    notes="Data returned to Tribal dept but no community-readable summary yet")
audit.respond("OCAP-A2", "no",
    notes="Technical report only — no plain language version produced")
audit.respond("OCAP-P1", "yes",
    notes="Tribal dept holds copy on local server")
audit.respond("OCAP-P2", "yes")

audit.respond("CARE-C1", "yes",
    notes="Benefit plan: drought early warning system for water managers")
audit.respond("CARE-C2", "partial",
    notes="Water managers consulted but broader community not involved")
audit.respond("CARE-A1", "yes")
audit.respond("CARE-A2", "no",
    notes="No formal mechanism to revoke sharing permissions")
audit.respond("CARE-R1", "yes")
audit.respond("CARE-R2", "yes",
    notes="Data sharing agreement covers permissible uses")
audit.respond("CARE-E1", "yes",
    notes="Reviewed — no culturally sensitive locations in gauge dataset")
audit.respond("CARE-E2", "partial",
    notes="Ethics documented for project period; post-project unclear")

audit.respond("FAIR-F1", "no",
    notes="No DOI assigned yet — pending Zenodo upload")
audit.respond("FAIR-F2", "yes")
audit.respond("FAIR-A1", "yes")
audit.respond("FAIR-I1", "yes",
    notes="GeoJSON + CSV with standard USGS field names")
audit.respond("FAIR-R1", "yes",
    notes="CC0 for public data; Tribal data gitignored")
audit.respond("FAIR-R2", "yes",
    notes="Provenance records in IEEE 2890-2025 format attached")

audit.respond("IEEE-1", "yes")
audit.respond("IEEE-2", "yes")
audit.respond("IEEE-3", "yes")
audit.respond("IEEE-4", "partial",
    notes="Derived datasets link to source but chain incomplete")
audit.respond("IEEE-5", "yes")

print(f"Answered: {len(audit.questions) - len(audit.unanswered)}/{len(audit.questions)}")
print(f"Overall score: {audit.overall_score:.0f}/100")
```

```python
# Print the full audit report
print(audit.report())
```

```python
# Save the audit report
from pathlib import Path
outputs = Path(REPO_ROOT) / "outputs"
outputs.mkdir(exist_ok=True)
audit.save(outputs / "sovereignty_audit.txt")
print("Saved: outputs/sovereignty_audit.txt")
```

## Interpreting the audit

**80–100:** Strong governance. Address remaining gaps before publication.

**60–79:** Adequate. Several gaps need attention: prioritize "no" responses
over "partial" responses.

**40–59:** Needs work. Consider pausing external sharing until key gaps
are resolved, particularly in OCAP® and CARE.

**Below 40:** Critical gaps. Do not publish or share externally until
governance is established.

**Note on "na" responses:** Questions marked not applicable (na) are
excluded from scoring. Use "na" sparingly and document why a question
does not apply as auditors may question this.
